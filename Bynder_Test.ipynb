## Exercise 1 

```python
medals_raw = pd.read_csv('D:\medals.csv')
#imported the csv file
```

```python
medals_raw.groupby(["Medal"]).agg({"Athlete":"nunique"})
#grouped the medals per category
```

```python
athlete_with_medals = medals_raw[medals_raw["Sport"] == "Athletics"].groupby(["Athlete"]).agg({"Gold":"sum", "Silver":"sum", "Bronze":"sum", "Event":"nunique"})
athlete_with_medals

athlete_with_gold = athlete_with_medals[athlete_with_medals["Gold"] != 0]
top_five_athlete_with_gold = athlete_with_gold.loc[:, ["Gold"]].sort_values(by="Gold", ascending=False).head(5)
top_five_athlete_with_gold

#created 'top_five_athlete_with_gold' dataframe with top 5 athletes 
```
